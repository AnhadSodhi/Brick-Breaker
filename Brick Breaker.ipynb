```python
%%ignite

circlex = 250
circley = 430

speedx = 0
speedy = -5

screenleft = bounding_box(0, 0, 5, 500)
screenright = bounding_box(499, 0, 5, 500)
screentop = bounding_box(0, 0, 500, 5)
screenbottom = bounding_box(0, 499, 500, 5)

boxnumber = 10

boxbound1 = bounding_box(0, 150, 30, 30)
boxbound2 = bounding_box(50, 150, 30, 30)
boxbound3 = bounding_box(100, 150, 30, 30)
boxbound4 = bounding_box(150, 150, 30, 30)
boxbound5 = bounding_box(200, 150, 30, 30)
boxbound6 = bounding_box(250, 150, 30, 30)
boxbound7 = bounding_box(300, 150, 30, 30)
boxbound8 = bounding_box(350, 150, 30, 30)
boxbound9 = bounding_box(400, 150, 30, 30)
boxbound10 = bounding_box(450, 150, 30, 30)

boxbound1hit = 0
boxbound2hit = 0
boxbound3hit = 0
boxbound4hit = 0
boxbound5hit = 0
boxbound6hit = 0
boxbound7hit = 0
boxbound8hit = 0
boxbound9hit = 0
boxbound10hit = 0

def setup():
    size(500,500)
    
def bar():
    fill_style("white")
    rect(mouse_x-35, 450, 70, 20)

def box(boxnumber):
    boxx = 0
    boxy = 150
    fill_style("red")
    for i in range(boxnumber):
        rect(boxx, boxy, 30, 30)
        boxx += 50
    
def ball():
    global speedx, speedy, circlex, circley
    
    circlex += speedx
    circley += speedy
    
    fill_style("yellow")
    circle(circlex, circley, 20)
    
def draw():
    global boxnumber, screenleft, screenright, screentop, screenbottom, speedx, speedy
    global boxbound1, boxbound2, boxbound3, boxbound4, boxbound5, boxbound6, boxbound7, boxbound8, boxbound9, boxbound10
    global boxbound1hit, boxbound2hit, boxbound3hit, boxbound4hit, boxbound5hit, boxbound6hit, boxbound7hit, boxbound8hit, boxbound9hit, boxbound10hit
    background("black")
    
    box(boxnumber)
    bar()
    
    ballbound = bounding_box(circlex, circley, 20, 20)
    barbound1 = bounding_box(mouse_x, 450, 14, 20)
    barbound2 = bounding_box(mouse_x+14, 450, 14, 20)
    barbound3 = bounding_box(mouse_x+28, 450, 14, 20)
    barbound3 = bounding_box(mouse_x+42, 450, 14, 20)
    barbound4 = bounding_box(mouse_x+56, 450, 14, 20)
    barbound5 = bounding_box(mouse_x+70, 450, 14, 20)
    
    if collided(barbound1, ballbound):
        speedx = -2.5
        speedy = -2.5
    if collided(barbound2, ballbound):
        speedx = -1.5
        speedy = -3.5
    if collided(barbound3, ballbound):
        speedx = 0
        speedy = -5
    if collided(barbound4, ballbound):
        speedx = 1.5
        speedy = -3.5
    if collided(barbound5, ballbound):
        speedx = 2.5
        speedy = -2.5
        
        
    if collided(screenleft, ballbound):
        speedx *= -1
    if collided(screenright, ballbound):
        speedx *= -1
    if collided(screentop, ballbound):
        speedy *= -1
    if collided(screenbottom, ballbound):
        speedy *= -1
    
    
    if collided(boxbound1, ballbound):
        boxbound1hit = 1
    if collided(boxbound2, ballbound):
        boxbound2hit = 1
    if collided(boxbound3, ballbound):
        boxbound3hit = 1
    if collided(boxbound4, ballbound):
        boxbound4hit = 1
    if collided(boxbound5, ballbound):
        boxbound5hit = 1
    if collided(boxbound6, ballbound):
        boxbound6hit = 1
    if collided(boxbound7, ballbound):
        boxbound7hit = 1
    if collided(boxbound8, ballbound):
        boxbound8hit = 1
    if collided(boxbound9, ballbound):
        boxbound9hit = 1
    if collided(boxbound10, ballbound):
        boxbound10hit = 1
        
        
    if boxbound1hit == 1:
        fill_style("black")
        rect(0, 150, 30, 30)
    if boxbound2hit == 1:
        fill_style("black")
        rect(50, 150, 30, 30)
    if boxbound3hit == 1:
        fill_style("black")
        rect(100, 150, 30, 30)
    if boxbound4hit == 1:
        fill_style("black")
        rect(150, 150, 30, 30)
    if boxbound5hit == 1:
        fill_style("black")
        rect(200, 150, 30, 30)
    if boxbound6hit == 1:
        fill_style("black")
        rect(250, 150, 30, 30)
    if boxbound7hit == 1:
        fill_style("black")
        rect(300, 150, 30, 30)
    if boxbound8hit == 1:
        fill_style("black")
        rect(350, 150, 30, 30)
    if boxbound9hit == 1:
        fill_style("black")
        rect(400, 150, 30, 30)
    if boxbound10hit == 1:
        fill_style("black")
        rect(450, 150, 30, 30)
        
    ball()
    
    boxhits = boxbound1hit + boxbound2hit + boxbound3hit + boxbound4hit + boxbound5hit + boxbound6hit + boxbound7hit + boxbound8hit + boxbound9hit + boxbound10hit
    if boxhits == 10:
        fill_style("black")
        rect(0, 0, 500, 500)
        image("you_win.jpg", 0, 0, 500, 500)
        fill_style("white")
        text("Credits", 200, 380)
        text("Coding: Anhad", 200, 400)
        text("Visual Design: Anhad", 200, 420)
        text("Technical Support: Afshin (Mentor)", 200, 440)
```
